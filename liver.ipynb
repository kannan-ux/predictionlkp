```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['Gender']=le.fit_transform(X['Gender'])
```

```python
X.drop('Gender',axis=1,inplace=True)
```

```python
X['Albumin_and_Globulin_Ratio'].fillna(X['Albumin_and_Globulin_Ratio'].mean(),inplace=True)
```

```python
import seaborn as sns

sns.boxplot(X['Age'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2,train_size=0.8)
```

```python
X_train
```

```python
X.head(356)
```

```python
X.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
liver=LogisticRegression()
```

```python
liver.fit(X_train,Y_train)
```

```python
Y_pred=liver.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
full_report=classification_report(Y_pred,Y_test)
```

```python
print(full_report)
```
