```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt




Q1 = X['MDVP:Fhi(Hz)'].quantile(0.25)
Q3 = X['MDVP:Fhi(Hz)'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

median_value = X['MDVP:Fhi(Hz)'].median()


X['MDVP:Fhi(Hz)'] = np.where((X['MDVP:Fhi(Hz)'] < lower_bound) | (X['MDVP:Fhi(Hz)'] > upper_bound), median_value, X['MDVP:Fhi(Hz)'])


sns.boxplot(X['MDVP:Fhi(Hz)'])
plt.show()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2,train_size=0.8)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train
```

```python
Y_train
```

```python
Y_pred=model.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
fullreport=classification_report(Y_pred,Y_test)
```

```python
print(fullreport)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train,Y_train)
train_score = dtree.score(X_train, Y_train)

#make predictions
Y_predict = dtree.predict(X_test)
```

```python
print(classification_report(Y_predict,Y_test))
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier

# Initialize and train a model
model = RandomForestClassifier()
model.fit(X_train, Y_train)

# Make predictions
y_pred_r = model.predict(X_test)
```

```python
print(classification_report(y_pred_r,Y_test))
```

```python
import xgboost as xgb
boost_model = xgb.XGBClassifier(
    n_estimators=100,   # Number of trees (boosting rounds)
    learning_rate=0.7,  # Learning rate
    max_depth=4,        # Max depth of each tree
    random_state=42
)

# Train the XGBoost model on the training data
boost_model.fit(X_train, Y_train)
Y_predictXg = boost_model.predict(X_test)
```
